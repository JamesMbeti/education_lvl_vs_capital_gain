```python
names = [ "age","workclass","fnlwgt","education","education-num","marital-status","occupation","relationship","race","sex","capital-gain","capital-loss","hours-per-week","native-country", "amount"]
```

```python
import pandas as pd
df = pd.read_csv("~/Downloads/adult.data", names = names)
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("darkgrid")
sns.set_context("talk")
barplot = sns.barplot(x = "education", y = "capital-gain", data = df, hue = "sex")
barplot.set(xlabel = "Education Levels", ylabel = "Capital Gain", title = "Capital Gain and Education level")
plt.xticks(rotation = 90);
```
